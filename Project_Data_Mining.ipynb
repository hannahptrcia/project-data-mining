Membersihkan data yang sudah diambil dengan mengubah semua huruf menjadi huruf kecil, menghapus RT, emoji, URL, hashtag, pengulangan huruf dan tanda baca

```python
import re
import emoji
import contractions

emoji_pattern = re.compile("["
        u"\U00002700-\U000027BF"  # Dingbats
        u"\U0001F600-\U0001F64F"  # Emoticons
        u"\U00002600-\U000026FF"  # Miscellaneous Symbols
        u"\U0001F300-\U0001F5FF"  # Miscellaneous Symbols And Pictographs
        u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        u"\U0001F680-\U0001F6FF"  # Transport and Map Symbols
                      "]+", re.UNICODE)

def cleansing(tweet):
    # Replace RT tag
    t1 = re.sub('RT\s', '', tweet)
    # Replace @_username
    t2 = re.sub('\B@\w+', "", t1)
    # Remove emojis
    t3 = emoji_pattern.sub(r'', t2)
    # Replace URL (http:// or https://)
    t4 = re.sub('(http|https):\/\/\S+', '', t3)
    # Replace #_something_
    t5 = re.sub('#+', '', t4)
    # Lower case each letter
    t6 = t5.lower()
    # Replace word repetition with a single occurance ('ooooooooo' becomse 'oo')
    t7 = re.sub(r'(.)\1+', r'\1\1', t6)
    # Replace punctuation repetition with a single occurance ('!!!!!!!!' becomes '!')
    t8 = re.sub(r'[\?\.\!]+(?=[\?.\!])', '', t7)
    # Alphabets only, exlude numbers and special characters
    t9 = re.sub(r'[^a-zA-Z]', ' ', t8)
    # Replace contractions with their extended forms
    t10 = contractions.fix(t9)
    return t10
```

```python
for i, r in df.iterrows():
    y = cleansing(r['full_text'])
    df.loc[i, 'full_text'] = y
df
```

Tokenizing atau memecah kalimat menjadi bagian-bagian kata

```python
def tokenize_text(text):
    tokenized_text = text.split()
    return tokenized_text
df['tokens'] = df['Tweet'].apply(lambda x: tokenize_text(x))
df.head()
```

Stopwords atau menghapus kata-kata yang tidak mengandung makna

```python
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

# ----------------------- get stopword from NLTK stopword -------------------------------
# get stopword indonesia
list_stopwords = stopwords.words('indonesian')


# ---------------------------- manualy add stopword  ------------------------------------
# append additional stopword
list_stopwords.extend([
    'yg','utk','cuman','deh','btw','tapi','gua','gue','gw', 'lo', 'lu',
    'kalo','klo','trs','jd','nih','ntar','entar','nya','lg','lgi','gk',
    'dpt','dr','kpn','kok','kyk','donk','dong','yah','u', 'ya', 'ga','gak',
    'km','eh','sih','bang','kyk','kan','gpp','usah','mas','sbb','thx','ato',
    'jg','dri','tp','tpi', 'haha', 'duh','ye','wkwk','wkwkwk','haha', 'iy', 'syg',
    'gaes','guys','lbh','loh','yuk','sdh','sy','dgn','tdk','kpd','tuh','dah','bbrp',
    'dlu','tuh', 'jadi', 'aja','bukan'
])
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]
df['remove_stopwords'] = df['tokens'].apply(stopwords_removal).copy()
df.head()
```

```python
import string
def remove_punct(text):
  text = " ".join([char for char in text if char not in string.punctuation])
  return text
df['clean'] = df['remove_stopwords'].apply(lambda x: remove_punct(x))
df
```

Freqdist atau melihat distribusi frekuensi dari masing-masing kata yang terdapat dari teks

```python
from nltk.probability import FreqDist
merged_list = []
for lst in df['remove_stopwords']:
    merged_list += lst
freq = FreqDist(merged_list)
freq
```

Memvisualisasikan hasil FreqDist

```python
import matplotlib.pyplot as plt
freq.plot(15, cumulative=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
text = " ".join(str(Tweet) for Tweet in df.clean)

wordcloud = WordCloud(max_font_size = 160, margin=0, background_color = "white", colormap="Greens").generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
